```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('out.csv')
```

```python
def bit_num_parse(x):
    try:
        return int(x)
    except:
        return np.nan
df['bit_num'] = df['bit_num'].apply(bit_num_parse)

df['sold'] = df['bit_num'].apply(lambda x: 0 if pd.isnull(x) else 1)
```

```python
df.head()
```

```python
size_cat_sdfs = []
for cat, sdf in df[['category', 'price', 'sold']].groupby(by=['category']):
    size = len(sdf)
    size_cat_sdfs.append( (size, cat, sdf))

maxed = max(size_cat_sdfs, key=lambda x:x[0])
```

```python
maxed[2]
```

```python
targets = []
for index, (size, cat, sdf) in enumerate(size_cat_sdfs):

    print(index, len(size_cat_sdfs))
    target = sdf.copy()

    mean = target['price'].mean()

    target['price_weight'] = target['price'].apply(lambda x:x/mean)

    def make_bin(price_weight):
        return price_weight - price_weight % 0.10
    target['price_weight'] = target['price_weight'].apply(make_bin)
    target = target[['sold', 'price_weight']]
    targets.append(target)
    
    #if index> 1000 :
    #    break
```

```python
targets = pd.concat([t for t in targets], axis=0)
targets
```

```python
for price_weight, sdf in targets.groupby(by=['price_weight']):
    shippings = sdf['sold'].tolist()
    if len(shippings) < 100:
        continue
    print(price_weight, sum(shippings)/len(shippings))
```
